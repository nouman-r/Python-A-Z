```python
# Distribution: 
vis1 = sns.distplot(stats['InternetUsers'])
vis1 = sns.distplot(stats['InternetUsers'], hist_kws=dict(edgecolor="k", linewidth=1)) # set bar border colors etc.
```

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y = 'BirthRate', fit_reg= False, hue = 'IncomeGroup', size=8,
                 scatter_kws = {'s':50})
# if we look at the help of sns.lmplot using shift+tab; it has dictionaries scatter_kws and line_kws.
# since seaborn is based on top of matplotlib we can lmplot is a wrapper function over matplotlib.pyplot.scatter. 
# there are few arguments that are not defined in sns.lmplot but are still in matplotlib.pyplot.scatter and 
# dict are used to accessing those. Like in case of marker size 's'
```
