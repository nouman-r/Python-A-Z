```python
# Define the style
sns.set(style="darkgrid", palette="muted", color_codes=True)

f, axes = plt.subplots()
f.set_size_inches(11.7, 8.27)
# Plot the boxsplots
ax = sns.boxplot(data=mov2, x='Genre', y='GrossUS', orient='v', color='lightgray', showfliers=False)
plt.setp(ax.artists, alpha=0.5)

# Add in points to show each observation
ax2 = sns.stripplot(x='Genre', y='GrossUS', data=mov3, jitter=True, size=8, edgecolor='black',\
                    linewidth=0, hue = mov3.Studio, alpha=0.7)

ax.axes.set_title('Domestics Gross % by Genre',fontsize=30)
ax.set_xlabel('Genre',fontsize=20)
ax.set_ylabel('Gross % US',fontsize=20)

# Define where to place the legend
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=1, prop={'size':20}, frameon=False,\
          facecolor='white')
```
